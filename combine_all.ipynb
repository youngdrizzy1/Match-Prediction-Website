```python
import pandas as pd
import glob
import os

print("="*70)
print("âš½ COMBINE SEASON FILES")
print("="*70)

season_files = sorted(glob.glob("pl_season_*.csv"))

if not season_files:
    print("\nâŒ No season files found!")
    print("   Looking for: pl_season_*.csv")
    exit()

print(f"\nğŸ“ Found {len(season_files)} file(s):")

all_data = []
for file in season_files:
    df = pd.read_csv(file)
    season = file.replace("pl_season_", "").replace(".csv", "")
    print(f"  âœ… {season}: {len(df)} matches")
    all_data.append(df)

combined_df = pd.concat(all_data, ignore_index=True)

print(f"\nğŸ“Š Before removing duplicates: {len(combined_df)} matches")

combined_df = combined_df.drop_duplicates(subset=['MatchID'], keep='first')

print(f"âœ… After removing duplicates: {len(combined_df)} matches")
print(f"ğŸ† Teams: {combined_df['Team'].nunique()}")
print(f"ğŸ“… Seasons: {combined_df['Season'].nunique()}")

combined_df.to_csv("premier_league_matches.csv", index=False)
print(f"\nğŸ’¾ Saved: premier_league_matches.csv")

print(f"\nğŸ“ˆ Matches per season:")
print(combined_df.groupby('Season').size().sort_index())

print("\n" + "="*70)
print("âœ… DONE! Now run: python train_model.py")
print("="*70)
```
